## 5. REGRESIÓN LINEAL



Se realizará una regresión lineal múltiple, siendo el precio de cierre de MELI la variable dependiente, y el resto de las evaluadas las independientes (predictores)

RECORDEMOS:

Una regresión lineal simple se basa en una sola variable independiente $X$, una variable dependiente $y$ y supone que existe una relación lineal entre estas. La ecuación general para este tipo de regresión es la siguiente:$$y = β_0 + β_1X + e$$

+ $β_0$ es la altura a la que la recta corta al eje Y, la cual se denomina término independiente.

+ $β_1$ es el incremento que se produce en la variable $y$ cuando la variable $X$ aumenta una unidad, el cual se denomina pendiente de la recta.

5.1. HIPÓTESIS

Se definen las hipótesis para contrastar con el análisis de regresión.

H0= Los predictores seleccionados no explican a la  linealmente a la variable dependiente (MELI) (B1=0)

H1= Los predictores seleccionados si explican linealmente a la variable dependiente (MELI) (B1=!0)

Creamos dos dataframe. Uno llamado "dfY" que va a contener los datos de la variable a predecir (MELI) y el otro llamado "dfX" que contiene los predictores o variables independientes.


```python
dfY= pd.DataFrame () #Creo un nuevo data frame
dfY["MELI"]=dfMELI.MELI #agrego columnas a este df

dfX= pd.DataFrame ()
dfX["META"]=archivo.META
dfX["GOOGLE"]=archivo.GOOGLE
dfX["TESLA"]=archivo.TESLA
dfX["APPLE"]=archivo.APPLE
dfX["AMAZON"]=archivo.AMAZON
dfX["TWITTER"]=archivo.TWITTER
dfX["PAYPAL"]=archivo.PAYPAL
dfX["GLOBANT"]=archivo.GLOBANT
dfX["DOLAR"]=archivo.DOLAR
dfX["MERVAL"]=archivo.MERVAL
dfX["RIESGO PAÍS"]=archivo.RIESGOPAIS
dfX["NASDAQ"]=archivo.NASDAQ
```

Seguidamente, importo las librerias necesarias para poder realizar la regresión OLS

```python
import matplotlib.pyplot       as plt
import seaborn as sns
import statsmodels.api as sm
```

Ajusto por OLS

```python
mod = sm.OLS(dfY, dfX)

res = mod.fit() #aca hago el ajuste

print(res.summary())
```
